```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# MÉTODOS NUMÉRICOS DEFINIDOS
# -----------------------------

def biseccion(a, b, f, tol=1e-5, max_iter=100):
    """
    Método de Bisección para encontrar una raíz de f en el intervalo [a, b].

    Parámetros:
    - a (float): extremo inferior del intervalo
    - b (float): extremo superior del intervalo
    - f (func): función a evaluar
    - tol (float): tolerancia del método
    - max_iter (int): máximo número de iteraciones

    Retorna:
    - list: lista de aproximaciones por iteración
    """
    if f(a) * f(b) >= 0:
        raise ValueError("La función no cambia de signo en el intervalo dado.")

    iteraciones = []
    for i in range(max_iter):
        c = (a + b) / 2
        iteraciones.append(c)
        if abs(f(c)) < tol or (b - a) / 2 < tol:
            break
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return iteraciones

def graficar_funcion(f, a, b, raiz=None, titulo=""):
    """
    Genera una gráfica de la función f en el intervalo [a, b].

    Parámetros:
    - f (func): función a graficar
    - a (float): límite inferior
    - b (float): límite superior
    - raiz (float): raíz estimada a marcar en la gráfica (opcional)
    - titulo (str): título de la gráfica
    """
    x = np.linspace(a, b, 400)
    y = f(x)
    plt.figure()
    plt.plot(x, y, label='f(x)')
    plt.axhline(0, color='black', linestyle='--')
    if raiz is not None:
        plt.plot(raiz, f(raiz), 'ro', label=f'Raíz ≈ {raiz:.6f}')
    plt.title(titulo)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)
    plt.show()


# ------------------------------------
# EJERCICIO 2 – MÉTODO DE BISECCIÓN
# ------------------------------------


# Ejercicio 2.a) f(x) = x - 2^(-x) en [0, 1]
def f2a(x):
    return x - 2**(-x)

raices_2a = biseccion(0, 1, f2a, tol=1e-5)
graficar_funcion(f2a, 0, 1, raices_2a[-1], "Ejercicio 2.a – Bisección")

# Ejercicio 2.b) f(x) = 2x*cos(2x) - (x+1)^2
def f2b(x):
    return 2 * x * np.cos(2 * x) - (x + 1)**2

raices_2b1 = biseccion(-3, -2, f2b, tol=1e-5)
graficar_funcion(f2b, -3, -2, raices_2b1[-1], "Ejercicio 2.b – Bisección en [-3, -2]")

raices_2b2 = biseccion(-1, 0, f2b, tol=1e-5)
graficar_funcion(f2b, -1, 0, raices_2b2[-1], "Ejercicio 2.b – Bisección en [-1, 0]")
```
