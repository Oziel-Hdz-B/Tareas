```python
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------
# MÉTODOS NUMÉRICOS DEFINIDOS
# ------------------------------

def Newton(x0, f, df, tol=1e-5, max_iter=100):
    """
    Método de Newton-Raphson para encontrar una raíz de f dada su derivada df.

    Parámetros:
    - x0 (float): valor inicial
    - f (func): función a evaluar
    - df (func): derivada de la función
    - tol (float): tolerancia
    - max_iter (int): máximo número de iteraciones

    Retorna:
    - list: lista de aproximaciones por iteración
    """
    iteraciones = [x0]
    for i in range(max_iter):
        fx = f(x0)
        dfx = df(x0)
        if dfx == 0:
            raise ValueError("La derivada es cero. Método de Newton falla.")
        x1 = x0 - fx / dfx
        iteraciones.append(x1)
        if abs(x1 - x0) < tol:
            break
        x0 = x1
    return iteraciones

def graficar_funcion(f, a, b, raiz=None, titulo=""):
    """
    Genera una gráfica de la función f en el intervalo [a, b].

    Parámetros:
    - f (func): función a graficar
    - a (float): límite inferior
    - b (float): límite superior
    - raiz (float): raíz estimada a marcar en la gráfica (opcional)
    """
    x = np.linspace(a, b, 400)
    y = f(x)
    plt.figure()
    plt.plot(x, y, label='f(x)')
    plt.axhline(0, color='black', linestyle='--')
    if raiz is not None:
        plt.plot(raiz, f(raiz), 'ro', label=f'Raíz ≈ {raiz:.6f}')
    plt.title(titulo)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)
    plt.show()

# ---------------------------------
# EJERCICIO 9 – MÉTODO DE NEWTON
# ---------------------------------


# Ejercicio 9.a) f(x) = e^x + 2^(-x) + 2cos(x) - 6
def f9a(x):
    return np.exp(x) + 2**(-x) + 2*np.cos(x) - 6

def df9a(x):
    return np.exp(x) - np.log(2)*2**(-x) - 2*np.sin(x)

raices_9a = Newton(1.5, f9a, df9a, tol=1e-5)
graficar_funcion(f9a, 1, 2, raices_9a[-1], "Ejercicio 9.a – Newton")

# Ejercicio 9.b) f(x) = 2x*cos(2x) - x^2
def f9b(x):
    return 2 * x * np.cos(2 * x) - x**2

def df9b(x):
    return 2 * np.cos(2 * x) - 4 * x * np.sin(2 * x) - 2 * x

raices_9b_1 = Newton(2.5, f9b, df9b, tol=1e-5)
graficar_funcion(f9b, 2, 3, raices_9b_1[-1], "Ejercicio 9.b – Newton en [2, 3]")

raices_9b_2 = Newton(3.5, f9b, df9b, tol=1e-5)
graficar_funcion(f9b, 3, 4, raices_9b_2[-1], "Ejercicio 9.b – Newton en [3, 4]")

# Ejercicio 9.c) f(x) = e^x - 3x^2
def f9c(x):
    return np.exp(x) - 3 * x**2

def df9c(x):
    return np.exp(x) - 6 * x

raices_9c_1 = Newton(0.5, f9c, df9c, tol=1e-5)
graficar_funcion(f9c, 0, 1, raices_9c_1[-1], "Ejercicio 9.c – Newton en [0, 1]")

raices_9c_2 = Newton(4.0, f9c, df9c, tol=1e-5)
graficar_funcion(f9c, 3, 5, raices_9c_2[-1], "Ejercicio 9.c – Newton en [3, 5]")
```
