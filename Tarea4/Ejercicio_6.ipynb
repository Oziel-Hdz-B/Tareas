6. Usar la iteración de Punto Fijo para mostrar que la función $g(x)=\pi+0.5 \sin (x / 2)$ tiene un punto fijo en el intervalo $[0,2 \pi]$ con una tolerancia de $10^{-2}$. (Programar el ejemplo en específico con una gráfica de la función).

```python
import math
import matplotlib.pyplot as plt
import numpy as np

# --------------------------------------------
# Función g(x) = π + 0.5·sen(x/2)
# Esta es la función usada para el método de Punto Fijo
# --------------------------------------------
def g(x):
    return math.pi + 0.5 * math.sin(x / 2)

# --------------------------------------------
# Parámetros iniciales
# --------------------------------------------
x0 = 2              # Valor inicial dentro del intervalo [0, 2π]
tol = 10**(-2)      # Tolerancia = 10^-2 como lo indica el ejercicio
itermax = 100       # Número máximo de iteraciones permitidas

# --------------------------------------------
# Inicialización para el ciclo de iteración
# --------------------------------------------
k = 0               # Contador de iteraciones
xk = x0             # Valor actual
xk1 = g(xk)         # Primer valor siguiente

# --------------------------------------------
# Iteración del Método de Punto Fijo
# --------------------------------------------
while abs(xk - xk1) > tol:
    if k >= itermax:
        break       # Salir si se alcanzó el máximo de iteraciones
    xk = xk1        # Actualizar valor actual
    xk1 = g(xk)     # Calcular nuevo valor
    k = k + 1       # Incrementar el contador

# --------------------------------------------
# Mostrar resultados
# --------------------------------------------
print("RESULTADO: Punto fijo de g(x)")
print("xk =",xk1)
print("Iteraciones realizadas:", k)

# --------------------------------------------
# Gráfica de g(x), la identidad y el punto fijo
# --------------------------------------------
x_vals = np.linspace(0, 2 * math.pi, 300)     # Valores de x entre 0 y 2π
g_vals = [g(x) for x in x_vals]               # Evaluación de g(x)
y_identidad = x_vals                          # Línea y = x

# Crear la gráfica
plt.figure(figsize=(14, 10))
plt.plot(x_vals, g_vals, label="g(x) = π + 0.5·sin(x/2)", color="#6B8E23")
plt.plot(x_vals, y_identidad, label="y = x", linestyle="--", color="#BF00BF")
plt.scatter([xk1], [xk1], color="#E7D40A", label="Punto fijo", zorder=5)

# Configuración de la gráfica
plt.title("Gráfica del Punto Fijo")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```
